```python
import pandas as pd
```

```python
df = pd.read_csv('data.tsv', sep='\t', encoding='utf-8',error_bad_lines=False,index_col=False)
```

```python
df.head()
```

```python
def preprocess(ReviewText):
    ReviewText = ReviewText.str.replace("(<br/>)", "")
    ReviewText = ReviewText.str.replace('(<a).*(>).*(</a>)', '')
    ReviewText = ReviewText.str.replace('(&amp)', '')
    ReviewText = ReviewText.str.replace('(&gt)', '')
    ReviewText = ReviewText.str.replace('(&lt)', '')
    ReviewText = ReviewText.str.replace('(\xa0)', ' ')  
    return ReviewText
```

```python
df['REVIEW_TEXT'] = preprocess(df['REVIEW_TEXT'])
```

```python
from textblob import TextBlob
df['polarity'] = df['REVIEW_TEXT'].map(lambda text: TextBlob(text).sentiment.polarity).filter()
```

```python
df['polarity'].head()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color = 'white',
        max_words = 200,
        max_font_size = 40, 
        scale = 3,
        random_state = 42
    ).generate(str(data))

    fig = plt.figure(1, figsize = (20, 20))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize = 20)
        fig.subplots_adjust(top = 2.3)

    plt.imshow(wordcloud)
    plt.show()
    
# print wordcloud
show_wordcloud(df["REVIEW_TEXT"])
```
